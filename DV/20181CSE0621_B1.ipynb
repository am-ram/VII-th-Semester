```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings; warnings.simplefilter('ignore')
plt.style.use(['dark_background','seaborn-pastel','ggplot'])
```

```python
df = pd.read_csv(r'company_sales_data.csv')
df.head(3)
```

# Part - B <br>Question.1-A

```python
facecream = df['facecream'] 
facewash = df['facewash']
months  = df ['month_number'].tolist()

with plt.style.context('dark_background'):
  width = 0.45
  x = np.arange(len(df['month_number']))
  plt.figure(figsize=(12, 8))
  plt.title('Sales of FaceCream & FaceWash',size=20,color='yellow',pad=20)
  plt.bar([(x-width/2) for x in months], facecream, width, label='Face Cream Sales',linewidth=3,color='cyan')
  plt.bar([(x+width/2) for x in months], facewash, width, label='Face Cream Sales',linewidth=3,color='hotpink')
  plt.xlabel('Months',size=20,color='yellow',labelpad=20)
  plt.xticks(size=20,color='yellow')
  plt.ylabel('FaceCream & FaceWash Sales',labelpad=20,size=20,color='yellow')
  plt.yticks(size=20,color='yellow')
  plt.legend(loc='upper left')
  plt.xticks(months)
```
