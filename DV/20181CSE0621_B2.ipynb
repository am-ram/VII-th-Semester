```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings; warnings.simplefilter('ignore')
plt.style.use(['dark_background','seaborn-pastel','ggplot' ])
```

```python
df = pd.read_csv(r'company_sales_data.csv')
df.head(3)
```

```python
months  = list(df['month_number'])
facecream   = df ['facecream'].tolist()
facewash   = df ['facewash'].tolist()
tpaste = df ['toothpaste'].tolist()
soap   = df ['bathingsoap'].tolist()
shampoo   = df ['shampoo'].tolist()
moist = df ['moisturizer'].tolist()
c = ['cornflowerblue','hotpink','orchid','darkorange','plum','cyan']
with plt.style.context('dark_background'):
  plt.figure(figsize=(10,8))
  plt.plot([],[],color='cornflowerblue', label='face Cream', linewidth=5)
  plt.plot([],[],color='hotpink', label='Face wash', linewidth=5)
  plt.plot([],[],color='orchid', label='Tooth paste', linewidth=5)
  plt.plot([],[],color='darkorange', label='Bathing soap', linewidth=5)
  plt.plot([],[],color='plum', label='Shampoo', linewidth=5)
  plt.plot([],[],color='cyan', label='Moisturizer', linewidth=5)
  plt.stackplot(months, facecream, facewash, tpaste, soap, shampoo, moist, colors=c)
  plt.xlabel('Month Number',size=20)
  plt.ylabel('Quantity of Sales',size=20)
  plt.grid(False)
  plt.title('Sales Data of Products',size=30,pad=20)
  plt.legend(loc='upper left')
  plt.show()
```
